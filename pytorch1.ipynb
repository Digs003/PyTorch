```python
tensor_A=torch.tensor([[1,2],
                       [3,4],
                       [5,6]])
tensor_B=torch.tensor([[7,10],
                       [8,11],
                       [9,12]])
```

```python
tensor_B_T=tensor_B.T
tensor_B_T
```

```python
print(f"Multiplying:{tensor_A.shape}*{tensor_B_T.shape}<-inner dimension match\n")
print("Output:\n")
output=torch.matmul(tensor_A,tensor_B_T)
print(output)
```

Reshaping,Viewing,Stacking,Squeezing and Unsqueezing

```python
x=torch.arange(1.,10.)
x,x.shape
```

```python
x_reshaped=x.reshape(9,1)
x_reshaped
```

```python
z=x.view(9,1)
z
```

```python
#Changing z changes x as both share same memory
z[0]=5
z,x
```

```python
x_stacked=torch.stack([x,x,x,x],dim=1)#dim=0-->vstack
x_stacked
```

```python
x_squeezed=x_reshaped.squeeze()#removes all single dimensions
x_squeezed,x_squeezed.shape
```

```python
x_unsqueezed=x_squeezed.unsqueeze(dim=0)
x_unsqueezed,x_unsqueezed.shape
```

```python
#Permuting
x_original=torch.rand(size=(224,224,3))
x_permuted=x_original.permute(2,0,1)
print(f"Previous shape:{x_original.shape}")
print(f"New shape:{x_permuted.shape}")
```

Reproducibility

```python
random_tensor_A=torch.rand(3,4)
random_tensor_B=torch.rand(3,4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A==random_tensor_B)
```

```python
#Lets make some randombut reproducible tensors
RANDOM_SEED=42
torch.manual_seed(RANDOM_SEED)
random_tensor_C=torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D=torch.rand(3,4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C==random_tensor_D)

```

Putting tensors on the GPU
